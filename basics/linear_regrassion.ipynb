## Linear regression（回帰直線（線形回帰））

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x_items = np.array(
    [14,23,75,93,92,48,28,74]
)
y_items = np.array(
    [23,31,34,68,79,58,36,41]
)

sns.regplot(y_items, x_items)
```

```python
# Use sklean

from sklearn.linear_model import LinearRegression

# The explanatory variables should be ndarray 
# where shape is number of samples, number of features.
# expand_dims: https://datawokagaku.com/numpy_functions/
X = np.expand_dims(x_items, axis=-1)

# Shape for the objective variable can be number of samples.
y = y_items

reg = LinearRegression()
reg.fit(X, y)

print("b={}".format(reg.coef_))
print("a={}".format(reg.intercept_))
```
